```python
import pickle
```

```python
def read_pickle(file):
    """
    Reads a pickle file and returns the object.
    
    :param obj: The object to be read from the pickle file.
    :param file: The path to the pickle file.
    :return: The object read from the pickle file.
    """
    with open(f"../app/ml_model/{file}", 'rb') as f:
        obj = pickle.load(f)
    return obj
```

```python
bin_svc = read_pickle('svc_bin_2.pkl')
bin_tree = read_pickle('tree_bin_2.pkl')
bin_knn = read_pickle('knn_bin_2.pkl')

mult_svc = read_pickle('svc.pkl')
mult_tree = read_pickle('dec_tree.pkl')
mult_knn = read_pickle('knn.pkl')
```

```python
print('Binary SVC:', bin_svc.get_params())
print('Binary Tree:', bin_tree.get_params())
print('Binary KNN:', bin_knn.get_params())

print('Multiclass SVC:', mult_svc.get_params())
print('Multiclass Tree:', mult_tree.get_params())
print('Multiclass KNN:', mult_knn.get_params())
```
